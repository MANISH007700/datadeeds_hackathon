## We will use a Regression Model of RandomForest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
rfr = RandomForestRegressor()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size= 0.3,random_state =100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# training
print(rfr.fit(x_train , y_train))
```

## Use metrics mean_squared_error

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
print(mean_squared_error(rfr.predict(x_train) , y_train))
```

```python
print(mean_squared_error(rfr.predict(x_test) , y_test))
```

```python
#test data

print(r2_score(rfr.predict(x_test) , y_test))
```

```python
#train data

print(r2_score(rfr.predict(x_train) , y_train))
```
